```python
# Complete the functions defined in this cell


# Test 01: Play Right
def playRight(sequence,beats):
  # TO DO
    for j in beats:
        if j==1:
            temp = sequence[len(sequence)-1]
            for i in range(len(sequence)-1, -1, -1):
                sequence[i] = sequence[i-1]
            sequence[0] = temp

    return sequence


# Test 02: Discard Cards
def discardCards(cards,number):
  # TO DO
    i=0
    while i <=(len(cards)-1):
        if cards[i] == number:
            for j in range(i, len(cards)-1):
                cards[j] = cards[j+1]
            cards[len(cards)-1] = 0
            i-=1
        i+= 1
    return cards


#Test 03: Merge Lineup
def mergeLineup(pokemon_1, pokemon_2):
  # TO DO
  i = 0
  j = len(pokemon_2)-1
  arr = np.zeros(len(pokemon_1), dtype='i')
  while i<= (len(pokemon_1)-1):
    if pokemon_1[i] == None:
        pokemon_1[i] = 0
    if pokemon_2[j] == None:
        pokemon_2[j] = 0

    arr[i] = pokemon_1[i] + pokemon_2[j]
    i+= 1
    j-= 1

  return arr


# Test 04: Balance your Salami
def balanceSalami(salami):
  # TO DO
  for i in range(len(salami)):
    left_sum = 0
    right_sum = 0
    for j in range(0, i):
        left_sum += salami[j]
    for k in range(i, len(salami)):
        right_sum += salami[k]

    if left_sum == right_sum:
        return True
  return False


# Test 05: Protecc Salami
def protectSalami(salami):
  # TO DO
    visited = np.zeros((len(salami)), dtype=bool)
    count_arr = np.zeros((len(salami)), dtype=int)

    for i in range(len(visited)):
        visited[i] = False

    for i in range(len(salami)):
        if visited[i] == True:
            continue
        
        count = 1
        for j in range(i+1, len(salami)):
            if salami[i] == salami[j]:
                visited[j] = True
                count+= 1
        
        count_arr[i] = count
    
    for i in range(len(count_arr)):
        if count_arr[i]>1:
            for j in range(i+1, len(count_arr)-1):
                if count_arr[i] == count_arr[j]:
                    return True

    return False


# Test 05: Odd Even Wave
def waveYourFlag(arr):
  # TO DO
    new_arr = np.zeros(len(arr), dtype=int)
    
    if arr[0] % 2 == 0:
        new_arr[0] = arr[0]
    

        i = 1
        j = 0
        while i< len(arr):
            if new_arr[i-1]%2 == 0 and arr[j]%2 != 0:
                new_arr[i] = arr[j]
                i += 1
                j -= 2
            elif new_arr[i-1]%2!= 0 and arr[j]%2 == 0:
                new_arr[i] = arr[j]
                i += 1
                j += 1

            j += 1

    if arr[0] % 2 != 0:
        new_arr[0] = arr[0]

        i = 1
        j = 0
        while i< len(arr):
            if new_arr[i-1]%2 == 0 and arr[j]%2 != 0:
                new_arr[i] = arr[j]
                i += 1
                j +=1
            elif new_arr[i-1]%2!= 0 and arr[j]%2 == 0:
                new_arr[i] = arr[j]
                i += 1
                j -= 2

            j += 1

    
    return new_arr
                
        




```

```python
import math
import numpy as np
 

def mean(arr):
    sum = 0
    count = 0
    for i in arr:
        sum += i
        count += 1
    mean = sum/count
    return mean
def standard_deviation(arr):
    mean1 = mean(arr)
    arr2 = np.zeros(len(arr))
    sum= 0
    for i in range(len(arr)):
        var = arr[i]-mean1
        arr2[i] = var
    for i in arr2:
        sum += i*i
    var3 = sum/(len(arr2)-1)
    stdev =  math.sqrt(var3)
    return stdev
def diff(arr):
    mean1 = mean(arr)
    stdev1 = standard_deviation(arr)
    arr4 = np.zeros(len(arr))
    a = 0

    x = mean1 + 1.5*(stdev1)
    y = mean1 - 1.5*(stdev1)

    for i in range(len(arr)):
        if arr[i]>=x:
            arr4[a] = arr[i]
            a+=1
        if arr[i]<=y:
            arr4[a] = arr[i]
            a+= 1

    arr5 = np.zeros(a, dtype=int)

    for i in range(a):
        arr5[i] = arr4[i]

    return arr5


```
